Matplotlib Markers

```python
#Using keyword "marker"
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "o")
plt.show()
```

```python
#formatting the line and color
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,"o-.g")
#"o" is for the circular marke, "-." is for dashed/dotted line, while "g" is for the color green.
#If you leave out the line value in format parameter, no line will be plotted.
plt.show()
```

```python
#Setting the size of the marker to 15

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 15)
#ms means marker size
plt.show()

```

```python
#Creating an edge color of the marker
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 15,mec="r")
#mec means marker edge color
plt.show()

```

```python
#Creating a face color of the marker
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 15,mec="r",mfc="r")
#mfc means markerfacecolor
plt.show()


#ms=marker size
#mec=marker edge color
#mfc=marker face color
```

Matplotlib Labels and Title and setting  font properties aand location of title

```python
#Creating labels for x and y axeses using the xlabel and ylabel method & Title too

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

#defining how the font of the title and labels will look like
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'darkred','size':15}


plt.title("Product Budget",fontdict= font1)
#creating a title for the plot using title function 
plt.xlabel("Average product count",fontdict= font2)
plt.ylabel("Averaage Budget ($)",fontdict= font2)

plt.plot(x, y)

plt.show()
```

```python
#Changing the position of the title. It can be right, left or center. By default is it at the center
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Product Budget",loc="left")
#using the "loc" parameter in the title method
plt.xlabel("Average product count")
plt.ylabel("Averaage Budget ($)")

plt.plot(x, y)

plt.show()
```

Matplotlib adding Grid Lines 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Product Budget")

plt.xlabel("Average product count")
plt.ylabel("Averaage Budget ($)")

plt.plot(x, y)
#creating grid using the grid() function
plt.grid()

plt.show()
```

```python
#Specify Which Grid Lines to Display

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Product Budget")

plt.xlabel("Average product count")
plt.ylabel("Averaage Budget ($)")

plt.plot(x, y)

plt.grid(axis="x") # returning only grid line only for the x axis

plt.show()
```

```python
plt.title("Product Budget")

plt.xlabel("Average product count")
plt.ylabel("Averaage Budget ($)")

plt.plot(x, y)

plt.grid(axis="y") # returning only grid line only for the y axis

plt.show()
```

```python
#Setting Line Properties for the Grid

plt.title("Product Budget")

plt.xlabel("Average product count")
plt.ylabel("Averaage Budget ($)")

plt.plot(x, y)

plt.grid(color="red",ls=":",linewidth=1) 
#Setting color to red, linestyle to dotted and linewidth to 1

plt.show()
```

```python
#Creating Title

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([1, 8, 4, 15])

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title("Sales")
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([5, 10, 15, 20])

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("Profit Margin")
plt.plot(x,y)

plt.show()
```

```python
#Super Title


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([1, 8, 4, 15])

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title("Sales")
#giving a title to each plot
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([5, 10, 15, 20])

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("Profit Margin")
#giving a title to each plot
plt.plot(x,y)

plt.suptitle("MY BUSINESS")
#giving an entire title (suptitile) to the figure 
plt.show()
```

Matplotlib Pie Charts

```python
#By default the plotting of the first wedge or piece starts from the x-axis and moves counterclockwise
y = np.array([45, 25, 25, 5])

plt.pie(y)
#creating a pie chart using pie() fucntion and passing varible(s) in it as argument
plt.show()
```

Labels

```python
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
#creating a mylabels variable
plt.pie(y, labels = mylabels)
#passing mylabels as the label of the pie chart
plt.show() 
```

Start Angle

```python
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
#creating a mylabels variable
plt.pie(y, labels = mylabels, startangle=90)
#setting the startangle to angle of 90
plt.show() 
```

Explode

```python
#making one of the piece to stand out using explode parameter
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
myexplode=[0.2,0,0,0]
#creating myexplode variable
plt.pie(y, labels = mylabels, explode=myexplode)

plt.show() 
```

Shadow

```python
#Adding shadow to pie chart
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
myexplode=[0.2,0,0,0]

plt.pie(y, labels = mylabels, explode=myexplode, shadow=True)
#setting the shadow parameter to True

plt.show() 
```

Colors

```python
#specifying new colors to wedege or piece
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
mycolors=["blue","yellow","purple","#4CBF50"]
#creating mycolors variable
plt.pie(y, labels = mylabels, colors=mycolors)


plt.show() 
```

Legend

```python
#creating legend using legend () fucntion
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
mycolors=["blue","yellow","purple","#4CBF50"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

Legend with Header

```python
#creating a header for the legend 
y = np.array([45, 25, 25, 5])
mylabels = ["Yam", "Potatoes", "Snacks", "Rice"]
mycolors=["blue","yellow","purple","#4CBF50"]

plt.pie(y, labels = mylabels)
plt.legend(title="Four Foods:")
#passing title argument on the legend function  
plt.show() 
```
