```python
#creating a new column("quality_label"), and defining a range and allocating that particular range to a label.
red_wine["quality_label"]=red_wine["quality"].apply(lambda value: "low"
                                                    if value <=5 else "medium"
                                                    if value <=7 else "high")
```

```python
red_wine
```

```python
#Transforming these labels into categrical data type instead of simple string
red_wine["quality_label"] = pd.Categorical(red_wine["quality_label"],
categories=["low", "medium", "high"])
```

```python
pd.concat([red_wine.describe(),white_wine.describe()], axis=1, keys=["Red_Wine Statistics","White_Wine Statistics"]).round(2)
```
