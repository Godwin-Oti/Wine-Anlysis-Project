Heatmap for Red wine

```python

redwine_no_label = pd.read_csv("winequality-red.csv", sep=None)
whitewine_no_label_with_dup = pd.read_csv("winequality-white.csv", sep=None)
whitewine_no_label_without_dup=whitewine_no_label_with_dup.drop_duplicates()
whitewine_no_label=whitewine_no_label_without_dup


combined_wine = pd.concat([redwine_no_label, whitewine_no_label])
corr_matrix = combined_wine.corr()

plt.figure(figsize=(15, 8))

# Create the heatmap using Seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidth=.5)

# Optional: Customize the heatmap appearance
plt.title(' Combined Wine Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Display the heatmap
plt.show()
```
