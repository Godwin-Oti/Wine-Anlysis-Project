```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Hello World!")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head(10)
```

```python
sns.displot(tips["total_bill"])
```

```python
#distribution plot
```

```python
#
```

```python
sns.barplot(data=tips,x ="sex", y="total_bill", estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
##box plot
sns.boxplot(x="day", y="total_bill", data=tips,hue="smoker")
```

```python
#violin plot
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True, palette="Set1")
```

```python
#stripplot
sns.stripplot(x="day",y="total_bill", data=tips, jitter =True, palette="deep", hue= "smoker")
```

```python
#sns.set_theme(style="whitegrid")
#sns.set_theme(style="")

```

```python
#swamplot
sns.swarmplot(x="day",y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day",y="total_bill", data=tips, color="black")
```

```python
sns.violinplot(y="day", x="total_bill", data=tips, color="green")
sns.swarmplot(y="day",x="total_bill", data=tips, color="black")
```

```python
#factorplot
sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
```

```python
#Regression plots
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,hue="sex", scatter_kws={"s":100},markers=["o","v"])
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,col="sex", row="smoker")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,hue="sex", col= "day", aspect=0.5)
```
