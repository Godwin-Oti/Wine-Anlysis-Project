Heatmap for Red wine

```python
#loading red wine csv file
redwine_no_label_with_dup = pd.read_csv("winequality-red.csv", sep=None)
#dropping duplicates
redwine_no_label_without_dup=redwine_no_label_with_dup.drop_duplicates()
redwine_no_label=redwine_no_label_without_dup
#resetting index and inplacing it.
redwine_no_label.reset_index(inplace=True, drop=True)

#giving a variable for matrix correlation of the data set
corr_matrix = redwine_no_label.corr()

# Set up the matplotlib figure
plt.figure(figsize=(15, 8))
#returning the matrix in triangular shape
mask = np.triu(corr_matrix)

# Create the heatmap using Seaborn
sns.heatmap(corr_matrix, mask=mask, annot=True, vmax=1, vmin=-1, center=0, cmap='coolwarm', fmt=".2f", linewidth=.5,  cbar_kws={"shrink": .5}, square=True)

# Customizing the heatmap appearance
plt.title('Red Wine Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Display the heatmap
plt.show()
```

Heatmap for White wine

```python
#creating a function that is capable taking a data set comparing alot numerical features to a categorical feature

def barplot_chart_loop(data,x):

    for i,col in enumerate(data.columns):
        #looping through the columns of the dataset
        plt.figure(i)
        sns.barplot(x=x,y=col,data=data)
        #defining x, y and data 
    return
```
