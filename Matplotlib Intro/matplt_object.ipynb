Adjusting the spacing around subplots

```python
fig, axes=plt.subplots(2,2, sharex=True, sharey=True)
#sharing x and y axes
for i in range(2):  # using loop function 
  for j in range(2):
    axes[i,j].hist(np.random.standard_normal(500), bins=50, color="black",alpha=0.5)
fig.subplots_adjust(wspace=0,hspace=0)
```

Colors, Markers and Line Style

```python
fig=plt.figure()
ax=fig.add_subplot()
data= np.random.standard_normal(30).cumsum()
ax.plot(data,color="r", ls="dashed", marker="o")
```

```python
fig=plt.figure()
ax=fig.add_subplot()
data= np.random.standard_normal(30).cumsum()
ax.plot(data,color="black", ls="dashed", label="Default")
ax.plot(data,color="black", ls="dashed", drawstyle="steps-post", label= "steps-post")
ax.legend()
```

Tick,Labels and Legends

```python
fig,ax=plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks=ax.set_xticks([0,250, 500,750,1000])
labels=ax.set_xticklabels(["one","two","three","four","five",],rotation=30,fontsize=8)
```

```python
fig,ax=plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks=ax.set_xticks([0,250, 500,750,1000])
labels=ax.set_xticklabels(["one","two","three","four","five",],rotation=30,fontsize=8)
ax.set_title("My Matplotlib")
#adding title
ax.set_xlabel("Gross($)")
```
